## EDA With Red Wine Data

Data Set Information:

The two datasets are related to red and white variants of the Portuguese "Vinho Verde" wine.  Due to privacy and logistic issues, only physicochemical (inputs) and sensory (the output) variables are available (e.g. there is no data about grape types, wine brand, wine selling price, etc.).

These datasets can be viewed as classification or regression tasks. The classes are ordered and not balanced (e.g. there are many more normal wines than excellent or poor ones). Outlier detection algorithms could be used to detect the few excellent or poor wines. Also, we are not sure if all input variables are relevant. So it could be interesting to test feature selection methods.


Attribute Information:

Input variables (based on physicochemical tests):
- 1 - fixed acidity
- 2 - volatile acidity
- 3 - citric acid
- 4 - residual sugar
- 5 - chlorides
- 6 - free sulfur dioxide
- 7 - total sulfur dioxide
- 8 - density
- 9 - pH
- 10 - sulphates
- 11 - alcohol

Output variable (based on sensory data):
- 12 - quality (score between 0 and 10)

```python
#univariate,bivariate,multivariate analysis
sns.pairplot(df)
```

1. Colors by quality: Each scatterplot point will be colored according to its quality rating.

2. Cluster patterns: You'll see if high-quality wines share common feature values (e.g., higher alcohol, lower volatile acidity).

3. Distribution separation: Diagonal plots will show whether quality groups have distinct distributions.

```python
sns.pairplot(df, hue="quality", diag_kind="kde", palette="coolwarm")
plt.show()
```

```python
# 1. Get correlation with quality
corr = df.corr()["quality"].abs().sort_values(ascending=False)

# 2. Select top features (excluding 'quality' itself)
top_features = corr.index[1:6]  # top 5 most correlated features
print("Top correlated features:", top_features.tolist())

# 3. Subset the dataframe
df_top = df[top_features.tolist() + ["quality"]]

# 4. Pairplot with hue
sns.pairplot(df_top, hue="quality", diag_kind="kde", palette="coolwarm")
plt.show()
```
