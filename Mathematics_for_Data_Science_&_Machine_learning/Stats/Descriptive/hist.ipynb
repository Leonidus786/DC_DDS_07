## üîç **What Are Histogram Bins, Recap**

A **bin** represents a range of values:

* Example: a bin from **10 to 20** groups all data points that fall in this range.
* The **height** of the bar (bin) = number of data points in that range.

---

## ‚úÖ **How to Select the Right Bin Size**

### 1. **Sturges‚Äô Rule**

Good for **small to medium datasets.**

```python
k = ‚åàlog2(n) + 1‚åâ
```

> For your data (n = 15):
> `k = ‚åàlog2(15) + 1‚åâ = ‚åà3.9 + 1‚åâ = 5 bins`

---

### 2. **Square Root Rule**

Good **default rule**.

```python
k = ‚åà‚àön‚åâ
```

> For n = 15 ‚Üí `k = ‚åà‚àö15‚åâ = 4 bins`

---

### 3. **Rice Rule**

Provides **slightly more granularity**.

```python
k = ‚åà2 * n^(1/3)‚åâ
```

> For n = 15 ‚Üí `k = ‚åà2 * 2.466‚åâ = 6 bins`

---

### 4. **Freedman-Diaconis Rule** *(best for skewed or uneven data)*

Uses **interquartile range (IQR)** and is great when:

* You have **outliers**
* The data is **not normally distributed**

```python
Bin width = 2 * IQR / n^(1/3)
```

> Then:

```python
k = (max - min) / bin_width
```
