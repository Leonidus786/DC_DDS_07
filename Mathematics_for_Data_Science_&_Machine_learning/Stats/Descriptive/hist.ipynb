##  **What Are Histogram Bins, Recap**

A **bin** represents a range of values:

* Example: a bin from **10 to 20** groups all data points that fall in this range.
* The **height** of the bar (bin) = number of data points in that range.

---

##  **How to Select the Right Bin Size**

### 1. **Sturges Rule**

Good for **small to medium datasets.**

```python
k = log2(n) + 1
```

> For your data (n = 15):
> `k = log2(15) + 1 = 3.9 + 1 = 5 bins`

---

### 2. **Square Root Rule**

Good **default rule**.

```python
k = n
```

> For n = 15  `k = 15 = 4 bins`

---

### 3. **Rice Rule**

Provides **slightly more granularity**.

```python
k = 2 * n^(1/3)
```

> For n = 15  `k = 2 * 2.466 = 6 bins`


---


### 4. **Freedman-Diaconis Rule** *(best for skewed or uneven data)*

Uses **interquartile range (IQR)** and is great when:

* You have **outliers**
* The data is **not normally distributed**

```python
Bin width = 2 * IQR / n^(1/3)
```

> Then:

```python
k = (max - min) / bin_width
```

---

##  **Your Dataset (Histogram)**

```python
data = [1, 12, 14, 18, 24, 26, 30, 35, 36, 37, 40, 41, 42, 43, 50]
```

###  Recommended Bins:

| Rule              | Number of Bins                     |
| ----------------- | ---------------------------------- |
| Sturges Rule     | 5                                  |
| Square Root Rule  | 4                                  |
| Rice Rule         | 6                                  |
| FreedmanDiaconis | \~68 (estimate, depending on IQR) |

##  How to Visualize This in Python

Try 3 different bin sizes to compare:

```python
import matplotlib.pyplot as plt

data = [1, 12, 14, 18, 24, 26, 30, 35, 36, 37, 40, 41, 42, 43, 50]

plt.figure(figsize=(15, 4))

# Sturges Rule - 5 bins
plt.subplot(1, 3, 1)
plt.hist(data, bins=5, edgecolor='black')
plt.title("Sturges Rule (5 bins)")

# Square Root Rule - 4 bins
plt.subplot(1, 3, 2)
plt.hist(data, bins=4, edgecolor='black')
plt.title("Square Root Rule (4 bins)")

# Rice Rule - 6 bins
plt.subplot(1, 3, 3)
plt.hist(data, bins=6, edgecolor='black')
plt.title("Rice Rule (6 bins)")

plt.tight_layout()
plt.show()
```

##  Final Tips:

* Start with **Square Root Rule**, then adjust visually.
* **Zoom in/out** on patterns by changing bins.
* **Domain knowledge** (e.g., income groups, age brackets) can guide bin ranges.




````python
import matplotlib.pyplot as plt

data = [1, 12, 14, 18, 24, 26, 30, 35, 36, 37, 40, 41, 42, 43, 50]

plt.figure(figsize=(15, 4))

# Sturges Rule - 5 bins
plt.subplot(1, 3, 1)
plt.hist(data, bins=5, edgecolor='black')
plt.title("Sturges Rule (5 bins)")

# Square Root Rule - 4 bins
plt.subplot(1, 3, 2)
plt.hist(data, bins=4, edgecolor='black')
plt.title("Square Root Rule (4 bins)")

# Rice Rule - 6 bins
plt.subplot(1, 3, 3)
plt.hist(data, bins=6, edgecolor='black')
plt.title("Rice Rule (6 bins)")

plt.tight_layout()
plt.show()
````
